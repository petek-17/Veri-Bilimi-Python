### ‚ùì Soru: Cinsiyete g√∂re yolcu sayƒ±sƒ±nƒ± g√∂rselle≈ütirin. (√∂rn. countplot veya bar chart)

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")

sex=df['sex'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(sex.index,sex.values,color=['blue','pink'],width=0.5)
plt.title('Cinsiyete G√∂re Yolcu Sayƒ±sƒ±')
plt.xlabel('Cinsiyet')
plt.ylabel('Yolcu Sayƒ±sƒ±')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Y ekseni i√ßin grid √ßizgileri
plt.tight_layout() # Grafik d√ºzenini sƒ±kƒ±≈ütƒ±r
plt.show()


```

### ‚ùì Soru: "Survived" deƒüi≈ükeninin daƒüƒ±lƒ±mƒ±nƒ± √ßiziniz.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
sns.countplot(x='survived', data=df)

plt.title('Survived Deƒüi≈ükeninin Daƒüƒ±lƒ±mƒ±')
plt.xlabel('Survived (0 = Hayƒ±r, 1 = Evet)')
plt.ylabel('Ki≈üi Sayƒ±sƒ±')
plt.show()
```

## üî¥ ƒ∞leri Seviye

### ‚ùì Soru: "Age" deƒüi≈ükeninin √ßarpƒ±klƒ±k (skewness) ve basƒ±klƒ±ƒüƒ±nƒ± (kurtosis) hesaplayƒ±n ve yorumlayƒ±n.

```python
# Cevabƒ±nƒ±zƒ± buraya yazƒ±nƒ±z
from scipy.stats import skew, kurtosis
# Age s√ºtunundaki eksik deƒüerleri kaldƒ±r
age_data = df['age'].dropna()

# √áarpƒ±klƒ±k ve basƒ±klƒ±k hesapla
age_skew = skew(age_data)
age_kurt = kurtosis(age_data, fisher=True) # fisher=True ise normal daƒüƒ±lƒ±mda 0 verir

print(f"√áarpƒ±klƒ±k (Skewness): {age_skew:.2f}")
print(f"Basƒ±klƒ±k (Kurtosis): {age_kurt:.2f}")

#√ßarpƒ±klƒ±k 0.39 gen√ßlerden √ßok ya≈ülƒ±lar var.Ama √ßok belirgin deƒüil, daƒüƒ±lƒ±m genel olarak simetrik sayƒ±labilir.
#baskƒ±nlƒ±k 0.17 Ya≈ülarƒ±n √ßoƒüu ortalamaya yakƒ±n, ama biraz u√ß deƒüer de var.
```

### ‚ùì Soru: Seaborn pairplot ile sayƒ±sal deƒüi≈ükenleri inceleyin. Hayatta kalma durumunu hue parametresiyle ayƒ±rƒ±n.
