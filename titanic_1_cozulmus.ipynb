### Soru 7: YaÅŸÄ± 60'tan bÃ¼yÃ¼k olanlarÄ±n hayatta kalma oranÄ±nÄ± bulun.

```python
df.groupby(df["Age"]>60)["Survived"].mean()
```

```python
# YaÅŸÄ± 60'tan bÃ¼yÃ¼k olanlarÄ± filtrele
older_than_60 = df[df['Age'] > 60]

# Hayatta kalanlarÄ± (survived == 1) say
survived_count = older_than_60['Survived'].sum()

# Toplam 60 yaÅŸ Ã¼stÃ¼ yolcu sayÄ±sÄ±
total_older = older_than_60.shape[0]

# Hayatta kalma oranÄ±
if total_older > 0:
    survival_rate = survived_count / total_older
    print(f"60 yaÅŸ Ã¼stÃ¼ yolcularÄ±n hayatta kalma oranÄ±: {survival_rate:.2%}")
else:
    print("Veride 60 yaÅŸ Ã¼stÃ¼ yolcu yok.")
```
