### Soru 1: Cinsiyete göre yaş ortalamasını bulun.

```python
df.groupby("Sex")["Age"].mean()
```

### Soru 2: Sınıfa göre hayatta kalan kişi sayısını bulun.

### Soru 3: Hem cinsiyet hem sınıfa göre hayatta kalma oranlarını hesaplayın.

```python
df.groupby(["Sex","Pclass"])["Survived"].mean()
```

### Soru 4: Embarked değişkenine göre ortalama yaş ve ortalama ücret nedir?

```python
df.groupby("Embarked")[["Age","Fare"]].mean()
```

### Soru 5: Pclass gruplarına göre yaşın standart sapmasını bulun.

```python
df.groupby("Pclass")["Age"].std()
```

### Soru 6: Her `Embarked` limanından kaç yolcu olduğunu ve bunların yüzdelik dağılımını yazdırın.

```python
embarked_counts=df['Embarked'].value_counts(dropna=False)
embarked_per=df['Embarked'].value_counts(normalize=True,dropna=False)*100
```

```python
sonuc= pd.DataFrame({
    'Yolcu Sayısı': embarked_counts,
    'Yüzde (%)': embarked_per.round(2)
})
print(sonuc)
```

### Soru 7: Yaşı 60'tan büyük olanların hayatta kalma oranını bulun.

```python
df.groupby(df["Age"]>60)["Survived"].mean()
```

```python
# Yaşı 60'tan büyük olanları filtrele
older_than_60 = df[df['Age'] > 60]

# Hayatta kalanları (survived == 1) say
survived_count = older_than_60['Survived'].sum()

# Toplam 60 yaş üstü yolcu sayısı
total_older = older_than_60.shape[0]

# Hayatta kalma oranı
if total_older > 0:
    survival_rate = survived_count / total_older
    print(f"60 yaş üstü yolcuların hayatta kalma oranı: {survival_rate:.2%}")
else:
    print("Veride 60 yaş üstü yolcu yok.")
```

### Soru 8: `Sex`, `Pclass`, `Survived` için `groupby()` + `agg()` ile ortalama yaş ve ücret hesaplayın.

```python
df.groupby(["Pclass","Sex","Survived"]).agg({
    "Age":"mean",
     "Fare":"mean"
})
```

### Soru 3: Hangi `Pclass` grubunda kadınların hayatta kalma oranı en yüksektir?

```python
# Kodunuzu bur
female=df[df["Sex"]=="female"]
survival_rates = female.groupby('Pclass')['Survived'].mean()
max_rate = survival_rates.max()
max_class = survival_rates.idxmax()
print(max_class)
```

### Soru 4: Ortalama yaşın altında olan ve hayatta kalan erkeklerin sayısı kaçtır?

```python
# Kodunuzu buraya yazın
avg_age=df["Age"].mean()
survivors=df[
    (df['Sex']=='male') &
    (df['Survived']==1) &
    (df['Age'] < avg_age)
].shape[0]
print(survivors)
```

### Soru 5: `Cabin` değeri eksik olmayanların yaş ortalaması ile eksik olanların yaş ortalaması arasında fark var mı?

```python
# Kodunuzu buraya yazın

cabin_isnull=df[df["Cabin"].isnull()]
```

```python
cabin_notnull=df[df["Cabin"].notnull()]
```

```python
print("Cabin bilgisi olanların yaş ortalaması:", cabin_isnull['Age'].mean())
print("Cabin bilgisi olmayanların yaş ortalaması:", cabin_notnull['Age'].mean())
```
